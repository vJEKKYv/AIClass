## RSS: Residual Sum of Squares

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# b1:  ÌöåÍ∑ÄÏãù ùë¶ _predict = 2ùê±+1 Ïóê ÎåÄÌï¥ Îç∞Ïù¥ÌÑ∞ (1, 4), (2, 4), (3, 8)Ïùò ÏûîÏ∞®Ï†úÍ≥±Ìï©ÏùÑ Í≥ÑÏÇ∞ÌïòÏûê.
data = np.array([[1, 4], [2, 4], [3, 8]])
plt.plot(data[:, 0], data[:, 1], 'ro')
x = np.linspace(start=1, stop=3, num=3) # y=2x+1
y = 2 * x + 1
print(y)
plt.plot(x, y, 'g-')
y_predict = 2*(data[:,0]) + 1
e = data[:,1] - y_predict
sum_of_squares = np.dot(e,np.transpose(e))
print(sum_of_squares)
```

```python
# b2: ÌöåÍ∑ÄÏãù ùë¶_hat = ùë§ùê±+1 Ïóê ÎåÄÌï¥ ùë§Ïùò Í∞íÏùÑ Î≥ÄÌôîÏãúÌÇ§Î©∞ ÏûîÏ∞®Ï†úÍ≥±Ìï©ÏùÑ Í≥ÑÏÇ∞ÌïòÍ≥†, Í∑∏ÎûòÌîÑÎ°ú ÌëúÌòÑÌïòÏûê.
data = np.array([[1, 4], [2, 4], [3, 8]])
# Define a range of w values
w_range = np.linspace(start=-10, stop=10, num=100)
# Compute the sum of squares for each w
sum_of_squares = []
for w in w_range:
    y_hat = w*(data[:,0]) + 1
    e = data[:,1] - y_hat
    sum_of_squares.append(np.dot(e,np.transpose(e)))

plt.plot(w_range, sum_of_squares)
# Find the optimal w for the given data
optimal_w = w_range[np.argmin(sum_of_squares)]
print(optimal_w) # 2.121212121...
# Plot the optimal w and the sum of squares
plt.plot(optimal_w, np.min(sum_of_squares), 'ro')
plt.show()
```

## OLS: Ordinary Least Squares

```python
from sklearn.datasets import make_regression
import statsmodels.api as sm
```

```python
# c1
b = 100
x0, y, w = make_regression(n_samples=200, n_features=1, bias=b, noise=30, coef=True, random_state=1)

x = sm.add_constant(x0)
y = y.reshape(len(y), 1)

plt.plot(x0, y, 'g.')
print(f'original model: y = {w:0.2f}x + {b}')
```
