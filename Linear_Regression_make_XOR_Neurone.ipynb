```python
def AND(values):
    weights = np.array([1,1])
    bias = -1.1
    return np.where((np.sum(values*weights)+bias)>0,1,0)
def OR(values):
    weights = np.array([1,1])
    bias = -0.9
    return np.where((np.sum(values*weights)+bias)>0,1,0)
def NOT(values):
    if values ==1:
        return 0
    else:
        return 1
def XOR(values):
    return (OR(np.array([AND(np.array([values[0],NOT(values[1])])),AND(np.array([NOT(values[0]), values[1]]))])))

print(XOR(np.array([0,0])),"\n")
print(XOR(np.array([1,0])),"\n")
print(XOR(np.array([0,1])),"\n")
print(XOR(np.array([1,1])),"\n")
```
