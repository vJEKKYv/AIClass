```python
import os  
import sys                           #파일시스템 열기(mnist 데이터 가져오기)
sys.path.append(os.pardir)
import numpy as np                   #행렬 연산용
import matplotlib.pyplot as plt      #함수 띄우기용
from dataset.mnist import load_mnist #github(밑바닥 부터 시작하는 딥러닝)제공 파일
import time                          #학습시간 측정용
import datetime                      #학습시간 띄우기용
```

```python
#활성화 함수 객체 정의. Relu와 Sigmoid를 순전파와 역전파에 맞게 함수까지 정의. 
class Relu:
    def __init__(self):
        self.mask = None

    def forward(self, x):
        self.mask = (x <= 0)
        out = x.copy()
        out[self.mask] = 0

        return out

    def backward(self, dout):
        dout[self.mask] = 0
        dx = dout

        return dx
class Sigmoid:
    def __init__(self):
        self.out = None

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))  
        
    def forward(self, x):
        out = self.sigmoid(x)
        self.out = out
        return out

    def backward(self, dout):
        dx = dout * (1.0 - self.out) * self.out

        return dx
```

```python
#형변환 함수. 이전 mlp의 nn.Linear()와 같은 역할. 
class Affine:
    def __init__(self, W, b):
        self.W = W
        self.b = b
        
        self.x = None
        self.original_x_shape = None
        # 가중치와 편향 매개변수의 미분
        self.dW = None
        self.db = None

    def forward(self, x):
        self.original_x_shape = x.shape
        x = x.reshape(x.shape[0], -1)
        self.x = x

        out = np.dot(self.x, self.W) + self.b

        return out

    def backward(self, dout):
        dx = np.dot(dout, self.W.T)
        self.dW = np.dot(self.x.T, dout)
        self.db = np.sum(dout, axis=0)
        
        dx = dx.reshape(*self.original_x_shape)  # 입력 데이터 모양 변경(텐서 대응)
        return dx
```

```python
##nn.Softmax()와 같은 함수. 
class SoftmaxWithLoss:
    def __init__(self):
        self.loss = None # 손실함수
        self.y = None    # softmax의 출력
        self.t = None    # 정답 레이블(원-핫 인코딩 형태)

    def softmax(self, x):
        if x.ndim == 2:
            x = x.T
            x = x - np.max(x, axis=0)
            y = np.exp(x) / np.sum(np.exp(x), axis=0)
            return y.T 

        x = x - np.max(x) # 오버플로 대책
        return np.exp(x) / np.sum(np.exp(x))

    def cross_entropy_error(self,y, t):
        batch_size = y.shape[0]
        return -np.sum(np.log(y[np.arange(batch_size), t] + 1e-8)) / batch_size
    
    def forward(self, x, t):
        self.t = t
        self.y = self.softmax(x)
        self.loss = self.cross_entropy_error(self.y, self.t)
        
        return self.loss

    def backward(self, dout=1):
        batch_size = self.t.shape[0]
        dx = self.y.copy()
        dx[np.arange(batch_size), self.t] -= 1
        dx = dx / batch_size
        
        return dx
```

```python
# 경사강하법 객체 정의. SGD와 ADAM을 각각 만들고 업데이트 함수 구현. 
class SGD:

    def __init__(self, lr=0.01):
        self.lr = lr
        
    def update(self, params, grads):
        for key in params.keys():
            params[key] -= self.lr * grads[key] 
class Adam:

    def __init__(self, lr=0.001, beta1=0.9, beta2=0.999): # 기본 파리미터 설정. 논문에 나온 것과 같이 설정했다. 
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.iter = 0
        self.m = None
        self.v = None
        
    def update(self, params, grads): # 논문에 나온 업데이트 룰. 
        if self.m is None:
            self.m, self.v = {}, {}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)
        
        self.iter += 1
        lr_t  = self.lr * np.sqrt(1.0 - self.beta2**self.iter) / (1.0 - self.beta1**self.iter)         
        
        for key in params.keys():
            self.m[key] += (1 - self.beta1) * (grads[key] - self.m[key])
            self.v[key] += (1 - self.beta2) * (grads[key]**2 - self.v[key])
            
            params[key] -= lr_t * self.m[key] / (np.sqrt(self.v[key]) + 1e-8)
```

```python
# MLP 구현. 
class MultiLayerNet:
    def __init__(self, input_size, hidden_size_list, output_size,
                 activation='relu', weight_init_std='relu', weight_decay_lambda=0): #활성화 함수 입력 없으면 Relu로 함. 
        self.input_size = input_size
        self.output_size = output_size
        self.hidden_size_list = hidden_size_list
        self.hidden_layer_num = len(hidden_size_list)
        self.weight_decay_lambda = weight_decay_lambda
        self.params = {}

        # 가중치 초기화
        self.__init_weight(weight_init_std)

        # 계층 생성
        activation_layer = {'sigmoid': Sigmoid, 'relu': Relu}
        self.layers = {}
        for idx in range(1, self.hidden_layer_num+1):
            self.layers['Affine' + str(idx)] = Affine(self.params['W' + str(idx)],
                                                      self.params['b' + str(idx)])
            self.layers['Activation_function' + str(idx)] = activation_layer[activation]()

        idx = self.hidden_layer_num + 1
        self.layers['Affine' + str(idx)] = Affine(self.params['W' + str(idx)],
            self.params['b' + str(idx)])

        self.last_layer = SoftmaxWithLoss()
    #He와 Xavier를 사용하여 값 초기화
    def __init_weight(self, weight_init_std): #weight_init_std 디폴트 값은 Relu
        #dictionary 로 layer 선택
        all_size_list = [self.input_size] + self.hidden_size_list + [self.output_size]
        for idx in range(1, len(all_size_list)):
            scale = weight_init_std
            
            if str(weight_init_std).lower() in ('relu', 'he'):
                scale = np.sqrt(2.0 / all_size_list[idx - 1])  # ReLU를 사용할 때 he
                
            elif str(weight_init_std).lower() in ('sigmoid', 'xavier'):
                scale = np.sqrt(1.0 / all_size_list[idx - 1])  # sigmoid를 사용할 때 xavier
                
            self.params['W' + str(idx)] = scale * np.random.randn(all_size_list[idx-1], all_size_list[idx])
            self.params['b' + str(idx)] = np.zeros(all_size_list[idx])
    #순전파
    def predict(self, x):
        for layer in self.layers.values():
            x = layer.forward(x)

        return x
    #손실함수
    def loss(self, x, t):
        
        y = self.predict(x)

        weight_decay = 0
        for idx in range(1, self.hidden_layer_num + 2):
            W = self.params['W' + str(idx)]
            weight_decay += 0.5 * self.weight_decay_lambda * np.sum(W ** 2)

        return self.last_layer.forward(y, t) + weight_decay
    #정확도 반환
    def accuracy(self, x, t):
        y = self.predict(x)
        y = np.argmax(y, axis=1)
        accuracy = np.sum(y == t) / float(x.shape[0])
        return accuracy
    
    def gradient(self, x, t):
        # forward
        self.loss(x, t)

        # backward
        dout = 1
        dout = self.last_layer.backward(dout)

        layers = list(self.layers.values())
        layers.reverse()
        for layer in layers:
            dout = layer.backward(dout)

        # 결과 저장
        grads = {}
        for idx in range(1, self.hidden_layer_num+2):
            grads['W' + str(idx)] = self.layers['Affine' + str(idx)].dW + self.weight_decay_lambda * self.layers['Affine' + str(idx)].W
            grads['b' + str(idx)] = self.layers['Affine' + str(idx)].db

        return grads
```

```python
# 0. MNIST 데이터 읽기==========
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True)

train_size = x_train.shape[0]
batch_size = 2500
max_iterations = 400


# 1. 실험용 설정==========
optimizers = {}
optimizers['Adam'] = Adam()
optimizers['SGD'] = SGD()

networks = {}
train_loss = {}

for key in optimizers.keys():
    #networks[key] = MultiLayerNet(input_size=784, hidden_size_list=[100, 100, 100, 100],output_size=10, activation='sigmoid',weight_init_std='sigmoid')
    networks[key] = MultiLayerNet(input_size=784, hidden_size_list=[400,256],output_size=10)
    train_loss[key] = []    

# 2. 훈련 시작==========
start = time.time()
print("학습시작")
for i in range(max_iterations):
    batch_mask = np.random.choice(train_size, batch_size)
    x_batch = x_train[batch_mask]
    t_batch = t_train[batch_mask]
    
    for key in optimizers.keys():
        grads = networks[key].gradient(x_batch, t_batch)
        optimizers[key].update(networks[key].params, grads)
    
        loss = networks[key].loss(x_batch, t_batch)
        train_loss[key].append(loss)
    
    if(i+1) % 200 == 0:
        print( "===========" + " 반복 횟수 : " + str(i+1) + " ===========")
        for key in optimizers.keys():
            loss = networks[key].loss(x_batch, t_batch)
            accuracy = networks[key].accuracy(x_batch, t_batch)
            print(key + "'s 정확도:" + str(accuracy * 100) + "%")
            print(key + "'s 손실률:" + str(loss))
end = time.time()
sec = (end - start)
result = datetime.timedelta(seconds=sec)
print("\n\n===========" + " 걸린 시간 : " + str(result) + " ===========\n\n")

# 3. 그래프 그리기==========
markers = {"SGD": "o", "Adam": "D"}
x = np.arange(max_iterations)
for key in optimizers.keys():
    plt.plot(x, train_loss[key], label=key, linewidth = '0.5')
plt.xlabel("iterations")
plt.ylabel("loss")
plt.ylim(0, 3)
plt.legend()
plt.show()
```

```python
iterations = x_test.shape[0]
accnt = {'Adam':0,'SGD':0}
test = {}
# 4. 성능 확인==========
for iter in range(iterations):
    for key in accnt.keys():
        test[key] = networks[key].predict(x_test[iter:iter+1]).argmax()
        if test[key] == t_test[iter]:
            accnt[key] += 1
for key in accnt.keys():
    print(key+"'s test Accuracy :\t"+str(accnt[key]/iterations*100)+"%")
```
