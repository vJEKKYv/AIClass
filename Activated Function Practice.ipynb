## 시그모이드/ 하이퍼볼릭 탄젠트, 계단, 렐루

```python
#np.linspace(-10,-10,n) -> 10~10을 n개의 수로 나눠서 순서대로 배열 생성
#np.where(a>3,3,a) -> a>3일때 3출력, 아닐때 a 출력
#np.maximun(a,3) -> 둘 중 큰 것 출력
#타겟 예제
Array_A = np.linspace(-10,10,200)
```

```python
#sigmoid 함수
x = Array_A
y = 1/(1+np.exp(-(Array_A)))
plt.plot(x,y)
```

```python
#Tanh 함수
x = Array_A
y = (np.exp(Array_A)-np.exp(-(Array_A)))/(np.exp(Array_A)+np.exp(-(Array_A)))
plt.plot(x,y)
```

```python
#Binary step 함수
x = Array_A
y = np.where(Array_A<=0,0,1)
plt.plot(x,y)
```

```python
#ReLU
x = Array_A
y = np.where(Array_A<=0,0,Array_A)
plt.plot(x,y)
```

## One neuron

```python
#단순 함수 만들어서 연산하기. 
Array_A = np.arange(2)
Array_B = np.arange(2)
for i in range(2):
    for j in range(2):
        plt.subplot(2,3,1)
        if AND(Array_A[i], Array_B[j]):
            plt.scatter(Array_A[i], Array_B[j], color='red', marker='o')
        else:
            plt.scatter(Array_A[i], Array_B[j], color='yellow', marker='o')
        plt.subplot(2,3,2)
        if OR(Array_A[i], Array_B[j]):
            plt.scatter(Array_A[i], Array_B[j], color='red', marker='o')
        else:
            plt.scatter(Array_A[i], Array_B[j], color='yellow', marker='o')
        plt.subplot(2,3,3)
        if NAND(Array_A[i], Array_B[j]):
            plt.scatter(Array_A[i], Array_B[j], color='red', marker='o')
        else:
            plt.scatter(Array_A[i], Array_B[j], color='yellow', marker='o')
            
        
def AND(a,b):
    if a*b > 0:
        return 1
    else:
        return 0
def OR(a,b):
    if a>0 or b>0:
        return 1
    else:
        return 0
def NAND(a,b):
    if a*b <= 0:
        return 1
    else:
        return 0
```

```python
# 원-뉴런 연산하기. 
#AND  연산의 함수는 f(x) = -x+1.5, 위에 있으면 True,
#OR   연산의 함수는 f(x) = -x+0.5, 위에 있으면 True,
#NAND 연산의 함수는 f(x) = -x+1.5, 위에 있으면 False,
# True는 빵강, false는 검정
for i in range(2):
    for j in range(2):
        plt.subplot(2,3,1)
        if i > (-j+2.5):
            plt.scatter(Array_A[i][], color='red', marker='o')
        else:
            plt.scatter(Array_A[i], Array_B[j], marker='o')
            
```
